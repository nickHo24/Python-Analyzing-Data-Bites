
Data for this article is taken from https://www.kaggle.com/rtatman/animal-bites

Column information:
- bite_date: Date the victim was bitten
- SpeciesIDDesc: Species of the animal causing the bite (eg dog, cat, ...)
- BreedIDDesc: Breed if any (eg PIT BULL dog breed)
- GenderIDDesc: Gender of the animal
- color: Color of the animal
- vaccination_yrs: Number of years since the animal was last vaccinated
- vaccination_date: The last time the animal was vaccinated
- victim_zip: Victim's zipcode
- AdvIssuedYNDesc: Did the victim receive advice from a doctor
- WhereBittenIDDesc: Location of being bitten
- quarantine_date: The date the animal was confined, if any
- DispositionIDDesc: Action of the authorities with the animal
- head_sent_date: Date the animal was brought to the research room to be tested for rabies, if any
- release_date: Date the animal was released, if any
- ResultsIDDesc: Rabies test results, if any

a) In the `species` DataFrame, which dog breed is the most popular?

b) Among black dogs, which breed is the most popular?

```python
dog = species[species['SpeciesIDDesc'] == 'DOG']
breed_dog = dog['BreedIDDesc'].unique()
max = 0
breed = ''
for name in breed_dog:
  data = dog[dog['BreedIDDesc'] == name].shape[0]
  if data > max:
    max = data
    breed = name
print('The most popular dog breed is :',breed)
```

```python
dog1 = dog.dropna(subset = ['color'])
black_dog = dog1[dog1['color'].str.contains('BLACK') | dog1['color'].str.contains('BLk')]
black_dog
```

```python
breed_black_dog = black_dog.BreedIDDesc.unique()
max_black = 0
breed_black = ''
for name_black in breed_black_dog:
  data_black = black_dog[black_dog['BreedIDDesc'] == name_black].shape[0]
  if data_black > max_black:
    max_black = data_black
    breed_black = name_black
print('The most popular black dog breed is :',breed_black)
```

In the years 2010 to 2016, which __varieties__ of __species__ caused the most bites? Do not count rows with the value `NaN` or `UNKNOWN` in these two columns. Suggestion: Should be divided into several steps.

```python
# Filter data for the years 2010 to 2016
filtered_data = bites[(bites['bite_date'].dt.year >= 2010) & (bites['bite_date'].dt.year <= 2016)]

# Remove rows with NaN or UNKNOWN in 'SpeciesIDDesc' and 'BreedIDDesc'
filtered_data = filtered_data[filtered_data['SpeciesIDDesc'].notna() & (filtered_data['SpeciesIDDesc'] != 'UNKNOWN')]
filtered_data = filtered_data[filtered_data['BreedIDDesc'].notna() & (filtered_data['BreedIDDesc'] != 'UNKNOWN')]

# Count the combinations of species and variety
bite_counts = filtered_data.groupby(['SpeciesIDDesc', 'BreedIDDesc']).size()

# Find the most frequent combination
most_bites = bite_counts.idxmax()
most_bites_count = bite_counts.max()

(most_bites, most_bites_count)
```

a) How many animals were recorded when they were brought to the research lab?

b) Among these animals, what percentage do dogs account for?

```python
number_animal = bites[pd.notna(bites).head_sent_date]
number_animal.shape[0]
```

```python
dog_number = number_animal[number_animal['SpeciesIDDesc'].str.contains('dog', case=False, na=False)].shape[0]
print(round(100 / number_animal.shape[0] * dog_number))
```

```python
# a) Counting the animals brought to the research lab
lab_visits_count = bites['head_sent_date'].notna().sum()

# b) Calculating the percentage of dogs among these animals
dogs_in_lab = bites[bites['SpeciesIDDesc'] == 'DOG']['head_sent_date'].notna().sum()
percentage_of_dogs = (dogs_in_lab / lab_visits_count) * 100 if lab_visits_count > 0 else 0

(lab_visits_count, percentage_of_dogs)
```

a) How many dogs have recorded release dates?

b) Among them, how many were recorded to have been vaccinated for at least 1 year?

```python
# a) Counting dogs with recorded release dates
dogs_with_release_dates = bites[(bites['SpeciesIDDesc'] == 'DOG') & (bites['release_date'].notna())]
number_of_dogs_with_release_dates = dogs_with_release_dates.shape[0]

# b) Counting vaccinated dogs with at least 1 year since last vaccination
vaccinated_dogs_over_a_year = dogs_with_release_dates[dogs_with_release_dates['vaccination_yrs'] >= 1]
number_of_vaccinated_dogs_over_a_year = vaccinated_dogs_over_a_year.shape[0]

(number_of_dogs_with_release_dates, number_of_vaccinated_dogs_over_a_year)
```
